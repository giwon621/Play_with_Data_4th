 ### 위치명 --> 주소 / 상호 찾기

```python
# https://developers.naver.com/docs/search/local/
    
import requests
import json


def find_address(placename, target):
    
    # 위치
    
    # 네이버API 키, 패스워드 입력하기
    fpath = 'c:/informs/naver_map.txt'
    with open(fpath, 'r') as fname:
        client_map_id, client_map_secret = fname.readline().split()
    
    # 특정 지역 상호 찾기 : ex) 망원 OOO,  망원 OOOO 
    if target not in placename:
        placename = target + ' ' +placename
    url = "https://openapi.naver.com/v1/search/local.json?query=" + placename
    headers = {'X-Naver-Client-Id': client_map_id, 'X-Naver-Client-Secret': client_map_secret}

    html  = requests.get(url, headers = headers).text
    datas = json.loads(html)
    
    for data in datas['items']:
#         print(data)
        if target in data['address']:
            address = data['address']
            title = data['title'].replace('</b>','').replace('<b>','')
            return title,address
    
    return '알수없음','알수없음'
```

```python
target = '마포구 망원동'
placename = 'Zapangi'

find_address(placename, target)
```

```python
from tqdm import tqdm_notebook
import time
```

```python
loc_locs = []
titles = []

for location in tqdm_notebook(locations):
    title, add = find_address(location, target)
    titles.append(title)
    loc_locs.append(add)
    print(title, add, sep = ' / ')
    time.sleep(0.3)
```

###  주소 --> 위도/경도 찾기

```python
# 주소 --> 위도/경도 확인하기
import requests
import json

def get_geo_xy(address):
    
    # 주소 --> 위도/경도 찾기
    
    
    # 네이버API 아이디/패스워드 입력하기
    fpath = 'c:/informs/naver_geo.txt'
    with open(fpath, 'r') as fname:
        client_geo_id, client_geo_secret = fname.readline().split()

    url = "https://openapi.naver.com/v1/map/geocode?query=" + address 
    headers = {'X-Naver-Client-Id': client_geo_id, 'X-Naver-Client-Secret': client_geo_secret}

    html = requests.get(url, headers = headers).text
    try:
        data = json.loads(html)
#         print(data)
        data_geo = data['result']['items'][0]['point']
        x = data_geo['x']
        y = data_geo['y']
    except: return(0,0)   # 값을 찾을 수 없을때는 0,0 을 가지고 옴
    return (x,y)
```

```python
address = '서울특별시 마포구 망원동 400-2'
get_geo_xy(address)
```

```python
from folium.plugins import MarkerCluster
```

```python
y = datas_geo.geo_y.values
x = datas_geo.geo_x.values

locations = list(zip(y,x))
popups = list(datas_geo.title.values)
```

```python
icon_create_function = """\
function(cluster) {
    return L.divIcon({
    html: '<b>' + cluster.getChildCount() + '</b>',
    className: 'marker-cluster marker-cluster-large',
    iconSize: new L.Point(30, 30)
    });
}"""
```

```python

map_mw2 = folium.Map(location=[37.557376, 126.904376], zoom_start = 16)  #망원시장 중심

marker_cluster = MarkerCluster(
    locations=locations, popups=popups,
    name='Mangwon',
    overlay=True,
    control=True,
    icon_create_function=icon_create_function
)

marker_cluster.add_to(map_mw2)
folium.LayerControl().add_to(map_mw2)


for index, row in datas_geo.iterrows():
    if index < 5:
        col = 'red'
    else: col = 'blue'
    
    if index < 50:
        geoData = [row['geo_y'], row['geo_x']]
        folium.CircleMarker(geoData, radius=0.2*row['count'],color=col,fill_color=col, popup=row['title'], fill = True).add_to(map_mw2)

map_mw2.save('map_mangwon2.html')        
map_mw2
```
